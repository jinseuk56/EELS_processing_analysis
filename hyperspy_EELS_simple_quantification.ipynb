```python
%matplotlib nbagg
import hyperspy.api as hys
import numpy as np
import tkinter.filedialog as tkf
```

```python
hl = hys.load(tkf.askopenfilename(), signal_type = "EELS")
ll = hys.load(tkf.askopenfilename(), signal_type = "EELS")
print(hl)
print(ll)
```

```python
hl.plot()
```

```python
hl.remove_background()
```

```python
roi = hys.roi.SpanROI(left=630, right=670)
hl.plot()
roi.add_widget(hl, axes=["Energy loss"])
```

```python
hl_mn = hl.isig[roi].integrate1D(axis="Energy loss")
hl_mn.plot()
```
