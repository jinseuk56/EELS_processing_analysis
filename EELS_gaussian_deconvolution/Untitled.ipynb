```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Cambria'
```

```python
def gauss(x, a, c, sigma):
    return a*np.exp(-(x-c)**2/(2*sigma**2))
```

```python
x_range = np.arange(-10, 10, 0.1)

p1 = [1.0, 0.0, 1.0]
p1_gauss = gauss(x_range, p1[0], p1[1], p1[2])
p1_area = np.trapz(x_range, p1_gauss)

p2 = [0.7, 0.0, 1.0]
p2_gauss = gauss(x_range, p2[0], p2[1], p2[2])
p2_area = np.trapz(x_range, p2_gauss)

print(p2_area / p1_area)
```

```python
plt.figure()
plt.plot(x_range, p1_gauss)
plt.plot(x_range, p2_gauss)
plt.show()
```
